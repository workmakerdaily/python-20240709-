##### 그래프의 종류

** 막대 그래프 ** : 데이터를 카테고리별로 값을 막대로 표현 할 때 사용하는 그래프 형식 (수직 막대 그래프, 수평 막대 그래프)
> 카테고리별 데이터를 비교하고자 할 때 사용 (제품 판매량, 지역별 인구수)  

** 선 그래프 ** : 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식  
> 하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할 때 사용 (기온, 주식)

** 파이 차트 ** : 전체 데이터에 대한 각 요소의 비율을 원형으로 나타낼 때 사용하는 그래프 형식  
> 전체에 대한 요소의 비율을 표현하고자 할 때 사용 (투표 결과)  

** 산점도 ** : 두 변수 간의 상관관계를 표현할 때 사용하는 그래프 형식  
> 각 요소간 관계성을 파악할 때 사용 (키와 몸무게 상관관계 비교)  

** 히스토그램 ** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의 수를 막대로 표현하는 방식  
> 데이터의 분포를 비교할 때 사용 (시험 점수 분포)  

** 영역 그래프 ** : 선 그래프의 아래쪽 빈 공간을 채운 그래프  
> 시간의 흐름에 따라 데이터를 비교할 때 사용, 변화량을 강조하고자 할 때 사용  

** 박스 그래프 ** : 데이터의 분포를 나타낼 때 사용하는 그래프 형식  
> 최대값, 최소값, 사분위수로 데이터의 이상치를 판단할 때 사용  

### 선 그래프
가장 기본적인 그래프 형식, 시간의 흐름 및 순서에 따라서 데이터의 변화량을 보여주고자 할 때 사용  
plot 패키지의 plot() 메서드로 간단하게 생성 가능

````python
plt.plot([1, 2, 4, 8, 16])
plt.show()
````

````python
# 그래프의 제목 지정 title() 메서드 사용
# plot(x축에 대한 리스트, y축에 대한 리스트)
plt.title('x 축 추가')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16])
plt.show()
````

````python
# 차트 스타일 지정
# plot() 메서드의 세번째 매개변수로  
# 지정된 문자열 포멧으로 전달하여 스타일 지정가능
plt.title('c2-- 스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 'c2--')
plt.show()
````

````python
# 스타일 매개변수
# color (c): 선 색상 , linewidth (lw): 선 굵기 , linestyle (ls): 선 스타일
# marker: 마커 종류, markersize (ms): 마커 크기 
# markeredgecolor (mec): 마커 테두리 색상, markeredgewidth (mew): 마커 테두리 굵기
# markerfacecolor (mfc): 마커 내부 색상
plt.title('스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.show()
````

````python
# 차트 범위 지정
# xlim(), ylim() 메서드로 x축과 y축의 상한 하한을 지정할 수 있음
plt.title('차트 범위 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xlim(2019, 2025)
plt.ylim(3, 20)
plt.show()
````

````python
# 틱 지정
# xticks(), yticks() 메서드로 틱의 위치나 라벨 지정 가능
plt.title('틱 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.show()
````

````python
# 그리드
# grid() 메서드로 그리드를 표시할지 결정할 수 있음
plt.title('그리드')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.show()
````

````python
# 차트 겹치기
# plot() 메서드를 중복해서 호출하여 차트를 겹칠 수 있음
plt.title('겹쳐 그리기')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
        c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#889900')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.show()
````

````python
# 범례
# regend() 메서드를 통해 범례를 표시할 수 있음
# loc 매개변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능
plt.title('범례 표시')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='증가')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
        c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#889900', label='감소')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.legend(loc=9)
plt.show()
````

````python
# 각 틱 라벨
# xlabel(), ylabel() 메서드로 x축, y축의 라벨을 표시할 수 있음
plt.title('각 틱 라벨')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 
        c='#00998850', lw=5, ls=':', marker='*', ms=10, mfc='#009988', label='증가')
plt.plot([2020, 2021, 2022, 2023, 2024], [16, 8, 4, 2, 1], 
        c='#88990050', lw=5, ls=':', marker='*', ms=10, mfc='#889900', label='감소')
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.xlabel('연도')
plt.ylabel('가치')
plt.grid(True)
plt.legend(loc=9)
plt.show()
````
